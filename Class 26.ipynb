```python
import matplotlib.pyplot as plt
%matplotlib inline

x = range(1,15)
y = [1,4,6,8,4,5,3,2,4,1,5,6,8,7]

plt.stackplot(x,y,colors = 'k',alpha = 0.3)

plt.plot(x,y,color ='k',marker ='o')
plt.grid(True)
plt.show()
```
