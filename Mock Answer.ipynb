```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#Loading the dataset
df = pd.read_csv('titanic.csv')
```

```python
#How number of  rows and column 
print(df.shape)
```

```python
#What are the column names and data types
print(df.dtypes)
```

```python
#What is the average age of the passengers?
print(df['age'].mean())
```

```python
#How many passengers survived?
print(df['survived'].sum())
```

```python
#  How many males and females are in the dataset?
print(df['sex'].value_counts())

```

```python
#What is the most common embarkation point?
print(df['embarked'].value_counts())
```

```python
#Create a bar chart showing the number of male vs female passengers.
gender_counts = df['sex'].value_counts()
plt.figure(figsize=(6, 4))
sns.barplot(x=gender_counts.index, y=gender_counts.values, palette=['blue', 'black'])
plt.title('Number of Male vs Female Passengers')
plt.xlabel('Sex')
plt.ylabel('Number of Passengers')
plt.show()
```

```python
#Create a histogram showing the distribution of passenger ages.
plt.hist(df['age'],edgecolor='yellow')
plt.title("Distribution of Passenger Ages")
plt.xlabel("Age")
plt.ylabel("Number of Passengers")
plt.show()
```

```python
#Create a pie chart of survival (Survived vs Not Survived).
survival_counts = df['survived'].value_counts()
labels = ['not survived', 'survived']
plt.pie(survival_counts, labels=labels, autopct='%1.1f%%', startangle=90, colors=['brown','blue'])
plt.title("survival Distribution")
plt.axis('equal')
plt.show()
```

```python
#Create a bar chart showing the number of passengers in eachclass(Pclass)
class_counts = df['pclass'].value_counts()
plt.figure(figsize=(6, 4))
sns.barplot(x=class_counts.index, y=class_counts.values, palette=['black','blue','red'])
plt.title('Number of Passengers in Each Class')
plt.xlabel('Passenger Class (Pclass)')
plt.ylabel('Number of Passengers')
plt.show()
```
