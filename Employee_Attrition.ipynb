```python
#what is co-relation?
#Relationship between two variables and ranging from -1 to 1

#+1 : perfect positive corelation, if one variable increases, other variable 
#increases propoertinally 

#-1 : perfect negative corelation : if one variable is increasing , other variable decreases proportionally

#0 : corelation : two variables are not related to each other on linear manner

# positive co-relation : (greater than 0 upto +1)


#example : work hours - if employee is working for more hours --> evaluation points
#given by manager will increase(+ve corelation)

#employee is working for more hours but his salary is low --> less satisfaction(-ve corelation)
```

```python
#feature extraction:

#positively corelated 

visuals = ['dept','numberOfProjects','timeSpent.company','workAccident','promotionInLast5years','salary']
```

```python
#plot all the features vs left column of the dataset:

import matplotlib.pyplot as plt
import seaborn as sns 
```

```python
fig = plt.subplots(figsize=(10,15))

for p,q in enumerate(visuals):      #p is for indexing of graphs and q is for having values to the graphs
      plt.subplot(3,2,p+1)
      plt.subplots_adjust(hspace = 1.0)
      sns.countplot(x=q, data = data, hue='left')
      plt.xticks(rotation=90)
```
