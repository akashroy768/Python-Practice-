```python
#how to create a scatter plot
```

```python
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
a = [10,20,30,40,50,60,70,80]
b = [5,3,2,5,6,1,4,2]
x = [1,2,3,4,3,2,1,2]

plt.scatter(a,b,color='b')
plt.scatter(a,x,color='r')
plt.plot(a,b)
plt.plot(a,x)
plt.show()
```

```python
#customizations of scatter plot

import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
a = [10,20,30,40,50,60,70,80]
b = [5,3,2,5,6,1,4,2]
x = [1,2,3,4,3,2,1,2]

plt.scatter(a,b,c='r',s = 100,edgecolors='y',marker='o',alpha=0.7)
plt.scatter(a,x,c='k',s = 150,edgecolors='y',marker='d',alpha=1)
plt.legend(['b','x'],loc='best')
plt.title("Scatter Plot Demo")
plt.xlabel("X-Axis")
plt.ylabel("Y-Axis")
plt.grid(True)
plt.savefig('Scatter Plot.png')
plt.show()
```

```python
#how to create a histogram
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

age = [10,12,16,19,11,20,26,28,30,38,35,34,45,60,68,64,62,70,78,75,79,85,94,95]
plt.hist(age,bins=[0,20,40,60,80,100])
plt.show()
```

```python
#customing the histogram
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

numbers = [10,12,16,19,11,20,26,28,30,38,35,34,45,60,68,64,62,70,78,75,79,85,94,95]

plt.hist(numbers,bins=[0,20,40,60,80,100], edgecolor='#000000', color = '#FF2331')
plt.title("Histogram Demo")
plt.xlabel("Range of values")
plt.ylabel("Frequency of values")
plt.grid(True)
plt.savefig("Histogram Plot.pdf")
plt.show()
```

```python
#how to create a pie chart
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

animal = ['Dog','Cat','Wolf','Lion']
count = [50,45,60,80]

plt.pie(count, labels = animal)
plt.title("pie chart demo")
plt.show()
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

animal = ['Dog','Cat','Wolf','Lion']
count = [50,45,60,80]

plt.pie(count, labels = animal,autopct='%1.1f%%',shadow=True, startangle = 90,explode = (0,0.1,0,0))
plt.title("pie chart demo")
plt.show()
```

```python
#How to create a Doughnut chart
```

```python
import matplotlib.pyplot as plt
%matplotlib inline

group_names = ["GroupA","GroupB","GroupC"]
group_size = [20,30,50]
size_centre = [10]

a,b,c = [plt.cm.Blues,plt.cm.Reds,plt.cm.Greens]

pie1 = plt.pie(group_size,labels = group_names, autopct = '%1.1f%%',radius = 1.5, shadow = True,startangle = 90,colors=[a(0.5),b(1.0),c(0.5)])
pie2 = plt.pie(size_centre,radius = 1.0,colors = 'w')
plt.show()
```
